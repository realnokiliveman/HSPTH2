1. Write a decorator `n_times` which given a number `n` makes a function to execute `n` times.

```python
def n_times(n):
    def decorator(func):
        def wrapper(*arguments, **kwarguments):
            for _ in range(n):
                func(*arguments, **kwarguments)
        return wrapper
    return decorator

@n_times(2)
def greet():
    print('Hello!')

greet()
```

2. Write a decorator `once` which makes a function to execute only once and other times it just returns the result calculated on the first call.


```python
import random

def once(func):
    result = None
    has_run = False

    def wrapper(*arguments, **kwarguments):
        nonlocal result, has_run
        if not has_run:
            result = func(*arguments, **kwarguments)
            has_run = True
        return result

    return wrapper

@once
def generate_lucky_number():
    return random.randint(1, 100)

print(generate_lucky_number())
print(generate_lucky_number())
print(generate_lucky_number())
```
